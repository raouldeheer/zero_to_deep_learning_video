# First Deep Learning Model

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

Dit zijn de imports die nodig zijn voor wiskunde. Zoals numpy voor nummers en matlib voor grafieken.

```python
from sklearn.datasets import make_circles
```

Dit importeerd te make_circels functie.

```python
X, y = make_circles(n_samples=1000,
                    noise=0.1,
                    factor=0.2,
                    random_state=0)
```

Deze functie creert 1000 random gegenereerde punten in een cirkel vorm.

```python
X
```

Hier worden alle punten afgedrukt die zijn gegenereerd.

```python
X.shape
```

Hier wordt informatie over de vorm afgedrukt.

```python
plt.figure(figsize=(5, 5))
plt.plot(X[y==0, 0], X[y==0, 1], 'ob', alpha=0.5)
plt.plot(X[y==1, 0], X[y==1, 1], 'xr', alpha=0.5)
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.legend(['0', '1'])
plt.title("Blue circles and Red crosses")
```

Hiermee wordt de dataset in twee groepen gesorteerd en afgedrukt in een grafiek. 

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
```

Hier wordt tensorflow geimporteerd voor machinelearning.

```python
model = Sequential()
```

Dit maakt een sequentieel model aan. 

```python
model.add(Dense(4, input_shape=(2,), activation='tanh'))
```

Hier voegen we een laag toe met 4 neuronen die allemaal met de allemaal zijn verbonden die als activation tahn gebruikt.

```python
model.add(Dense(1, activation='sigmoid'))
```

Hier voegen we nog een output laag toe met een output neuron met de activatie sigmoid die met alle vorige is verbonden.

```python
model.compile(SGD(learning_rate=0.5), 'binary_crossentropy', metrics=['accuracy'])
```

Dit compileerd de neural network aan de hand van accuracy.

```python
model.fit(X, y, epochs=20)
```

Dit traint de neural network aan de hand van de training data met 20 iteraties van de trainingset data. 

```python
hticks = np.linspace(-1.5, 1.5, 101)
vticks = np.linspace(-1.5, 1.5, 101)
aa, bb = np.meshgrid(hticks, vticks)
ab = np.c_[aa.ravel(), bb.ravel()]
c = model.predict(ab)
cc = c.reshape(aa.shape)
```

Dit genereerd de gebieden om aan te geven wat de zekerheid is van welk type de dot zal zijn.

```python
plt.figure(figsize=(5, 5))
plt.contourf(aa, bb, cc, cmap='bwr', alpha=0.2)
plt.plot(X[y==0, 0], X[y==0, 1], 'ob', alpha=0.5)
plt.plot(X[y==1, 0], X[y==1, 1], 'xr', alpha=0.5)
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.legend(['0', '1'])
plt.title("Blue circles and Red crosses")
```

Dit tekend de grafiek met de punten en gebieden.
